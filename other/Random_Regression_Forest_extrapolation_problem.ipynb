# Random Regression Forest extrapolation problem

**Random Regression Forest** are composed by several Random Regression Tree.
The prediction of each Random Tree is averaged with the other prediction to produce a single result. However, we must beware of the extrapolation problem that may arise when using Random Regression Forest. This phenomenon appears when we try to predict a point that is outside the domain of the training dataset.

This notebook wants, in short, to show this phenomenon.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

## Dataset

```python
def dataset_f(x):
  return x ** 2
```

```python
# Creating vectors X and Y
x = np.linspace(-100, 100, 500)
y = dataset_f(x)

fig = plt.figure(figsize = (10, 5))
# Create the plot
plt.plot(x, y)
plt.legend(["Actual distribution"])
# Show the plot
plt.show()
```

```python
train_size = 50
x_train = np.random.randint(0,100,size=(train_size, 1))
y_train = []
for i in range(train_size):
  y_train.append( dataset_f(x_train[i])[0] )
```

## Train and Prediction

```python
RFmodel = RandomForestRegressor().fit(x_train, y_train)
```

```python
x_test = np.linspace(-100, 100, 500)
y_test = dataset_f(x_test)

prediction = RFmodel.predict(x_test.reshape(-1, 1))
```

## Final results

```python
fig = plt.figure(figsize = (10, 5))
# Create the plot

y = dataset_f(x_test)

plt.plot(x_test, y) # Actual function
plt.plot(x_test, prediction, 'r') # Random Forest Prediction
 
plt.legend(["Actual distribution", "Random Forest Prediction"])
# Show the plot
plt.show()
```
